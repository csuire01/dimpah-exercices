### vi. Chunking

This example shows a simple chunk grammar consisting of two rules. The first rule matches an optional determiner (`DT`) or possessive pronoun (`PP`), zero or more adjectives (`JJ`) then a noun (`NN`) or more nouns (`NNS`). The second rule matches one or more proper nouns (`NNP`). We will use the previous sentence with the already detected parts of speech, and run the chunker on this input.

```python
import nltk

grammar = r"""
  NP: {<DT|PP\$>?<JJ>*<NN|NNS>}   # chunk determiner/possessive, adjectives and noun or nouns
      {<NNS>+}                # chunk sequences of proper nouns
"""
chunker = nltk.RegexpParser(grammar) # chunker

chunks_sentence = chunker.parse(pos_tags_sentence)

print(chunks_sentence)
```
